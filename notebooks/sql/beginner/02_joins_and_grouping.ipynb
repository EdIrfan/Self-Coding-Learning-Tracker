## Q1 — Basic INNER JOIN

For each track, return the track's `Name` (as `TrackName`) and its album's `Title`
(as `AlbumTitle`), by joining `Track` to `Album` on `AlbumId`. Limit to 10 rows.

```python
query = """
-- your SQL here
SELECT 
    t1.Name AS TrackName,
    t2.Title AS AlbumTitle
FROM
    Track t1
INNER JOIN
    Album t2
ON 
    t1.AlbumId = t2.AlbumId
LIMIT 10
"""
run(query)
```

---
## Q2 — Three-table JOIN

Extend Q1: also bring in the artist's name (`Artist.Name`, as `ArtistName`) by joining
`Album` to `Artist` on `ArtistId`. Return `TrackName`, `AlbumTitle`, `ArtistName`, limit 10.

```python
query = """
-- your SQL here
SELECT 
    t1.Name AS TrackName,
    t2.Title AS AlbumTitle,
    t3.Name AS ArtistName
FROM
    Track t1
INNER JOIN
    Album t2
ON 
    t1.AlbumId = t2.AlbumId
INNER JOIN
    Artist t3
ON
    t2.ArtistId = t3.ArtistId
LIMIT 10
"""
run(query)
```

## Q3 — JOIN + GROUP BY + COUNT

Count how many tracks exist in each genre. Join `Track` to `Genre` **on `GenreId`**
(that's the shared key — neither table has a `GenreName` column; the genre's name is
just `Genre.Name`). Group by `Genre.Name`, return it aliased as `Genre` along with
`NumTracks`, ordered by `NumTracks` descending.

```python
query = """
-- your SQL here
SELECT
    t2.name AS Genre,
    COUNT(*) AS NumTracks
FROM
    Track t1
INNER JOIN
    Genre t2
ON
    t1.GenreId = t2.GenreId
GROUP BY
    t2.name
ORDER BY
    COUNT(*) DESC
"""
run(query)
```

---
## Q4 — GROUP BY + SUM (no join needed)

From the `Invoice` table, compute total amount spent per customer: `CustomerId` and
`TotalSpent` (sum of `Total`), ordered by `TotalSpent` descending, top 5 only.

```python
query = """
-- your SQL here
SELECT 
    CustomerId,
    sum(Total) AS TotalSpent
FROM
    Invoice
GROUP BY
    CustomerId
ORDER BY
    sum(Total) DESC
LIMIT 5
"""
run(query)
```

## Q5 — HAVING (filter on an aggregate)

Same setup as Q3 (tracks per genre), but this time only return genres with **more than
100 tracks**. Remember: `WHERE` can't filter on an aggregate — you need `HAVING`.

```python
query = """
-- your SQL here
SELECT
    t2.name AS GenreName,
    COUNT(*) AS NumTracks
FROM
    Track t1
INNER JOIN
    Genre t2
ON
    t1.GenreId = t2.GenreId
GROUP BY
    t2.name
HAVING
    COUNT(*) > 100
ORDER BY
    COUNT(*) DESC
"""
run(query)
```
