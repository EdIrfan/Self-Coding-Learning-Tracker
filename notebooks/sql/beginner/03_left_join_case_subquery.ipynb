## Q1 — LEFT JOIN + IS NULL (anti-join pattern)

Find customers who have **never placed an invoice**: `LEFT JOIN` `Customer` to `Invoice`
on `CustomerId`, keep rows where `Invoice.InvoiceId IS NULL`. Return `Customer.CustomerId`,
`FirstName`, `LastName`.

**Heads up:** in this dataset every customer has at least one invoice, so the verified
correct answer is genuinely **0 rows** (with the right 3 columns) — that's not a bug,
it's the whole point of the pattern (`LEFT JOIN` + `WHERE right.key IS NULL` finds rows
on the left with no match on the right). This exact pattern is a very common interview
question ("find customers with no orders") even when, on a given dataset, it's empty.

```python
query = """
-- your SQL here
SELECT 
    t1.CustomerId,
    t1.FirstName,
    t1.LastName
FROM
    Customer t1
LEFT JOIN
    Invoice t2
ON
    t1.CustomerId = t2.CustomerId
WHERE
    t2.InvoiceId IS NULL
"""
run(query)
```

## Q2 — DISTINCT

From the `Invoice` table, return each distinct `BillingCountry` value once, ordered
alphabetically.

```python
query = """
-- your SQL here
SELECT
    DISTINCT BillingCountry
FROM
    Invoice
ORDER BY
    BillingCountry ASC
"""
run(query)
```

## Q3 — MIN / MAX / AVG in one query

From the `Track` table, return one row with three columns: `MinMs` (minimum
`Milliseconds`), `MaxMs` (maximum `Milliseconds`), `AvgMs` (average `Milliseconds`).

```python
query = """
-- your SQL here
SELECT 
    min(Milliseconds) AS MinMs,
    max(Milliseconds) AS MaxMs,
    avg(Milliseconds) AS AvgMs
FROM
    Track
"""
run(query)
```

## Q4 — CASE WHEN + GROUP BY

From `Track`, bucket every track by `Milliseconds` using a `CASE WHEN`: **"Short"** if
`Milliseconds < 240000` (4 minutes), else **"Long"**, aliased as `LengthBucket`. Group by
that bucket and return `LengthBucket` with `NumTracks` (count of tracks in that bucket).

```python
query = """
-- your SQL here
SELECT
    CASE WHEN
        Milliseconds < 240000 THEN 'Short'
    ELSE 'Long' 
    END AS LengthBucket,
    COUNT(*) AS NumTracks
FROM
    Track
GROUP BY
    LengthBucket
"""
run(query)
```

## Q5 — Basic (non-correlated) subquery

From `Track`, return `Name` and `UnitPrice` for tracks priced **above the overall average
`UnitPrice`** across all tracks. Use a subquery — `WHERE UnitPrice > (SELECT AVG(UnitPrice)
FROM Track)` — rather than hardcoding the average. Limit to 10 rows.

```python
query = """
-- your SQL here
SELECT
    Name,
    UnitPrice
FROM
    Track
WHERE
    UnitPrice > (SELECT AVG(UnitPrice) FROM Track)
LIMIT 10
"""
run(query)
```
