```python
import sqlite3
import pandas as pd

conn = sqlite3.connect("../../../datasets/chinook.db")

def run(query: str) -> pd.DataFrame:
    return pd.read_sql_query(query, conn)

# sanity check — list the tables in this database
run("SELECT name FROM sqlite_master WHERE type='table' ORDER BY name")
```

---
## Q1 — Basic SELECT + LIMIT

Return every column from the `Genre` table, but only the first 10 rows.

```python
query = """
select * from Genre limit 10
"""
run(query)
```

---
## Q2 — WHERE with a numeric filter

From the `Track` table, return the `Name` and `UnitPrice` of tracks that are longer than 5 minutes.

(Hint: `Milliseconds` is the duration column — 5 minutes = 300000 ms.)

```python
query = """
-- your SQL here
SELECT
    Name,
    UnitPrice
FROM
    Track
WHERE
    Milliseconds > 300000
"""
run(query)
```

---
## Q3 — WHERE with a string filter + ORDER BY

From the `Customer` table, return customers where `Country = 'Canada'`, ordered alphabetically by `City`.

```python
query = """
-- your SQL here
SELECT
    *
FROM
    Customer
WHERE
    Country = 'Canada'
ORDER BY
    City

"""
run(query)
```

---
## Q4 — ORDER BY DESC + LIMIT (top-N)

Find the 5 most expensive tracks in the `Track` table (`Name` and `UnitPrice`).

```python
query = """
-- your SQL here
SELECT
    Name,
    UnitPrice
FROM
    Track
ORDER BY
    UnitPrice DESC
LIMIT 5
"""
run(query)
```

---
## Q5 — NULL handling

From the `Employee` table, return `FirstName`, `LastName`, and `Title` for employees who **do** report to someone (i.e. `ReportsTo` is not null).

```python
query = """
-- your SQL here
SELECT
    FirstName,
    LastName,
    Title
FROM
    Employee
WHERE
    ReportsTo IS NOT NULL
"""
run(query)
```
